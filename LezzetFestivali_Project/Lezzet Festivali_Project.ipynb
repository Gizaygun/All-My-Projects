```python
import pandas as pd
import numpy as np

katilimci_verileri = {
    'Isim': ["Ahmet", "Ayşe", "Beren", "Bora", "Cemre", "Cem", "Deniz", "Derya", "Ece", "Eray"],
    'Yas': [35, 30, 28, 24, 25, 20, 32, 22, 34, 25],
    'Cinsiyet': ["Erkek", "Kadın", "Kadın", "Erkek", "Kadın", "Erkek", "Erkek", "Kadın", "Kadın", "Erkek"],
    'Tercih_Edilen_Yemek': ["Beyti", "Kebap", "Şırdan", "Kokoreç", "Kebap", "Kokoreç", "Beyti", "Kebap", "Şırdan", "Kebap"],
    'Harcanan_Sure': [2, 3, 4, 2, 3, 1, 4, 4, 3, 2]

}
populer_etkinlikler = ["Ahşap Boyama", "Tiyatro Gösterisi", "Resim Sergisi", "Yöresel Danslar"]

while len(populer_etkinlikler) < len(katilimci_verileri['Isim']):
    populer_etkinlikler *= 2

katilimci_df = pd.DataFrame(katilimci_verileri)

katilimci_df['Populer_Etkinlik'] = populer_etkinlikler[:len(katilimci_verileri['Isim'])]

ortalama_harcanan_sure = np.mean(katilimci_df['Harcanan_Sure'])
medyan_yas = np.median(katilimci_df['Yas'])
toplam_harcanan_sure = np.sum(katilimci_df['Harcanan_Sure'])

populer_etkinlik_analiz = katilimci_df['Populer_Etkinlik'].value_counts().reset_index(name='Katilimci_Sayisi')

print("Veri Seti (DataFrame):")
print(katilimci_df)
print("\nOrtalama Harcanan Süre:", ortalama_harcanan_sure, "saat")
print("Medyan Yaş:", medyan_yas, "yaş")
print("Toplam Harcanan Süre:", toplam_harcanan_sure, "saat")
print("\nPopüler Etkinliklere Göre Analiz:")
print(populer_etkinlik_analiz)
```
